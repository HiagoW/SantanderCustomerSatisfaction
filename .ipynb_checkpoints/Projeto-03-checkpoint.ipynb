```python
# Cria array com nomes das colunas
headerArr = header[0].split(',')
headerArr
```

```python
rdd3 = rdd2.map(lambda line: line.split(','))
```

```python
# Função que cria objeto do tipo Row
def criaRow(line):
    d = {}
    for k,v in enumerate(line):
        d[headerArr[k]] = v
    return d
```

```python
df = rdd3.map(lambda x: Row(**criaRow(x))).toDF()
```

```python
# Converte tudo para double, exceto ID (Int)
df2 = df.select(
  *(
    F.col(col).cast("double").alias(col)
    if col != 'ID'
    else F.col(col).cast("int").alias(col)
    for col
    in df.columns
  )
)
```

```python
# Carrega o dict com as correlações
with open('corr.json') as json_file:
    corrDict = json.load(json_file)
```

```python
# Remove nan
corrDict = {k: v for k, v in corrDict.items() if str(v) != 'nan'}
```

```python
# Cria DF e ordena do maior pro menor
corrDF = pd.DataFrame.from_dict(corrDict, orient='index',columns=['corr'])
corrDF = corrDF.sort_values('corr')
```

```python
# Retirar 150 variáveis, as 75 menores e 75 maiores
bestCorrColumns = pd.concat([corrDF.iloc[0:75], corrDF.iloc[260:335]], axis=0)
```

```python
# Lista com as melhores colunas
bestColumns = bestCorrColumns.index.values.tolist()
bestColumns.append('TARGET')
```

```python
# Extrai melhores colunas
df3 = df2[bestColumns]
```

```python
len(df3.columns)
```

```python
df3.cache()
```

```python
# Criando um LabeledPoint (target, Vector[features])
def transformaVar(row) :
    obj = (row["TARGET"], Vectors.dense([row[c] for c in bestColumns]))
    return obj
```

```python
rddLabeled = df3.rdd.map(transformaVar)
```

```python
rddLabeled.collect()
```

```python
df4 = spSession.createDataFrame(rddLabeled,["label", "features"])
df4.select("features", "label").show(10)
```

```python
# Aplicando Redução de Dimensionalidade com PCA
rddPCA = PCA(k = 10, inputCol = "features", outputCol = "pcaFeatures")
pcaModel = rddPCA.fit(df4)
pcaResult = pcaModel.transform(df4).select("label","pcaFeatures")
pcaResult.show(truncate = False)
```

```python
# Dados de Treino e de Teste
# (dados_treino, dados_teste) = pcaResult.randomSplit([0.7, 0.3])
(dados_treino, dados_teste) = df4.randomSplit([0.7, 0.3])
```

```python
dados_treino.count()
```

```python
dados_teste.count()
```

```python
# Treinamento e criação do modelo
# linearReg = LinearRegression(maxIter = 10,featuresCol='pcaFeatures')
linearReg = LinearRegression(maxIter = 10)
modelo = linearReg.fit(dados_treino)
```

```python
# Imprimindo as métricas
print("Coeficientes: " + str(modelo.coefficients))
print("Intercepto: " + str(modelo.intercept))
```

```python
# Previsões com dados de teste
predictions = modelo.transform(dados_teste)
# predictions.select("pcaFeatures", "prediction").show()
predictions.select("features", "prediction", "label").show()
```

```python
# Coeficiente de determinação R2
avaliador = RegressionEvaluator(predictionCol = "prediction", labelCol = "label", metricName = "r2")
avaliador.evaluate(predictions) 
```
